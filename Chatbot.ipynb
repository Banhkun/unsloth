```python
import os
import gradio as gr
from langchain.vectorstores import FAISS
from google import genai
from google.genai import types

# Assuming you already have your documents and FAISS vector store created like so:
# documents = [Document(...), ...]
# vector_store = FAISS.from_documents(documents, embedding_model)

def generate_response_rag(user_input):
    # Retrieve top 10 most relevant documents
    retrieved_docs = vector_store.similarity_search(user_input, k=10)

    # Combine the retrieved context into a single string, including the title if available.
    # Also collect the titles for references.
    context_parts = []
    references = []
    for doc in retrieved_docs:
        if hasattr(doc, "title") and doc.title:
            # Append both the title and the content to the context
            context_parts.append(f"{doc.title}\n{doc.page_content}")
            references.append(doc.title)  # Store the title as a reference (assuming it contains a link)
        else:
            context_parts.append(doc.page_content)
    context_text = "\n\n".join(context_parts)
    print(context_text)

    # Construct the prompt with integrated context and question,
    # and instruct the model to provide references at the end.
    prompt = (
        "You are an AI assistant with expert knowledge. Answer the following question strictly based on the provided information.\n\n"
        f"Context:\n{context_text}\n\n"
        f"Question: {user_input}\n\n"
        "🔹 **Instructions:**\n"
        "1️⃣ **Follow the structure of the context exactly** (maintain the order of text and images as presented).\n"
        "2️⃣ **If an image link is provided, render the image out** (do not modify, shorten, or remove links).\n"
        "3️⃣ **Do NOT add information beyond the provided context** (if uncertain, explicitly state missing information).\n"
        "4️⃣ **At the end of your answer, include a 'References' section that lists the titles (with exact links) of the documents used.**\n\n"
    )

    # Set up the Gemini client
    client = genai.Client(api_key=os.environ.get("GEMINI_API_KEY"))
    model = "gemini-2.0-flash-thinking-exp-01-21"

    # Build the content with our augmented prompt
    contents = [
        types.Content(
            role="user",
            parts=[types.Part.from_text(text=prompt)]
        )
    ]

    # Define the generation configuration
    generate_content_config = types.GenerateContentConfig(
        temperature=0.7,
        top_p=0.95,
        top_k=64,
        max_output_tokens=65536,
        response_mime_type="text/plain",
    )

    # Stream the response from the Gemini model
    full_response = ""
    for chunk in client.models.generate_content_stream(
            model=model,
            contents=contents,
            config=generate_content_config,
    ):
        full_response += chunk.text

    # Append a References section based on the retrieved document titles
    references_text = "\n".join(f"- {ref}" for ref in references)
    full_response_with_refs = f"{full_response}\n\nReferences:\n{references_text}"
    return full_response_with_refs

# Chat function that maintains conversation history
def chat(user_message, history):
    if history is None:
        history = []
    bot_response = generate_response_rag(user_message)
    history.append((user_message, bot_response))
    return history, history

# Build the Gradio interface
with gr.Blocks() as demo:
    chatbot = gr.Chatbot(height=800)
    msg = gr.Textbox(placeholder="Enter your message here...", label="Your Message")
    state = gr.State([])

    # Update the chat history when the user submits a message
    msg.submit(chat, inputs=[msg, state], outputs=[chatbot, state])

demo.launch(share=True)
```
